# Month-2 Data Analysis Portfolio Summary

This notebook provides a high-level overview of all five projects completed during the Month-2 Data Analysis Internship. It highlights the domain, objective, and key insight from each project.

## Project-01 — Supermarket Sales (Retail)
**Objective:** Analyze revenue trends and customer behavior.

**Key Insight:** Member customers spend more per transaction and peak sales occur during specific hours.

## Project-02 — Student Performance (Education)
**Objective:** Understand how academic performance varies across subjects and attendance.

**Key Insight:** Higher attendance strongly correlates with better academic performance.

## Project-03 — Weather Data Analysis
**Objective:** Analyze temperature, rainfall, and extreme weather patterns.

**Key Insight:** Seasonal variations significantly impact temperature and rainfall distribution.

## Project-04 — COVID-19 Healthcare Trends
**Objective:** Study infection, recovery, and mortality trends.

**Key Insight:** COVID-19 showed multiple waves with sharp growth and decline patterns.

## Project-05 — Social Media Analytics
**Objective:** Analyze engagement patterns across posts and users.

**Key Insight:** Engagement varies significantly based on posting time and content type.
